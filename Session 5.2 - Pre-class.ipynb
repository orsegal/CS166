## Exercise 15.2

* The number of edges in a complete graph made of $n$ nodes
    * Each node is connected to $n-1$ nodes, and every edge is shared by 2 nodes so we divide by 2:
    * $N_{edge} = \frac{n(n-1)}{2}$
* The number of edges in a regular graph made of $n$ nodes each of which has degree $k$.
    * Similar to before, just that each node is now connected to $k$ edges:
    * $N_{edge} = \frac{nk}{2}$

## Exercise 15.3

```python
from pylab import *
import networkx as nx
```

```python
# A complete graph made of four nodes
g = nx.Graph()

g.add_nodes_from([1, 2, 3, 4])
g.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)])

nx.draw(g)
show()
```

```python
# A complete graph made of five nodes
g = nx.Graph()

g.add_nodes_from([1, 2, 3, 4, 5])
g.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5), (4, 5)])

nx.draw(g)
show()
```

```python
# Bipartite - two and three
g = nx.Graph()

g.add_nodes_from([1, 2, 3, 4, 5])
g.add_edges_from([(1, 4), (1, 5), (2, 4), (2, 5), (3, 4), (3, 5)])

color_map = ["red"]*3 + ["blue"]*2
nx.draw(g, node_color = color_map, with_labels = True)
```

```python
# Bipartite - three and three
g = nx.Graph()

g.add_nodes_from([1, 2, 3, 4, 5, 6])
g.add_edges_from([(1, 4), (1, 5), (1, 6), (2, 4), (2, 5), (2, 6), (3, 4), (3, 5), (3, 6)])

color_map = ["red"]*3 + ["blue"]*3
nx.draw(g, node_color = color_map, with_labels = True)
```

## Exercise 15.3
Every connected tree graph made of $n$ nodes has exactly $n-1$ edges. Explain why.

If we start looking from the leafs, we will see that evey leaf is connected to one edge going up towards the root.
That is true for every node apart from the root itself that is not connected upwards, hence $n-1$.

```python
g = nx.balanced_tree(2, 3)
nx.draw(g)
show()
```
